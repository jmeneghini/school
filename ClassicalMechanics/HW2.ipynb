# Fourier Series
$x(t) = -1$ if $t < 0$ \
$x(t) = 1$ if $t > 0$

```python
def fourier(t, num_terms) :
    """
    Returns the Fourier series approximation of a square wave
    with the given number of terms.
    """
    f = 0
    for n in range(1, num_terms + 1) :
        f += 4/np.pi * np.sin((2*n - 1)*t)/(2*n - 1)
    return f    
```

```python
fig, ax = plt.subplots(figsize=plots.set_size('default', fraction=1))

t = np.linspace(-np.pi, np.pi, 1000)

# plot the function
ax.plot(t, fourier(t, 2), label='2 terms')
ax.plot(t, fourier(t, 3), label='3 terms')
ax.plot(t, fourier(t, 4), label='4 terms')

ax.set_xlabel('t')
ax.set_ylabel('x(t)')

ax.legend()

fig.savefig('fourier.png', dpi=300)
```
