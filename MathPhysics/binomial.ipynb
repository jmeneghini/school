```python
from scipy.special import factorial
import numpy as np
import matplotlib.pyplot as plt
import scienceplots

plt.style.use(['science'])
# set default figure size
plt.rcParams['figure.figsize'] = [8,6]
```

```python
def binomial(n, k, a):
    return factorial(n) / (factorial(k) * factorial(n - k)) * a**k * (1 - a)**(n - k)
```

```python
# plot binomial distribution for different values of n
a = 0.5
n_values = [10, 20, 30, 40]
k_values = np.arange(0, 40)

for n in n_values:
    plt.plot(k_values, binomial(n, k_values, a), label=f'n={n}')
    
plt.xlabel('k')
plt.ylabel('P(X=k)')
plt.legend()
plt.title(f'Binomial distribution for different values of n\na={a}')
```

```python
a = 0.25

for n in n_values:
    plt.plot(k_values, binomial(n, k_values, a), label=f'n={n}')

plt.xlabel('k')
plt.ylabel('P(X=k)')
plt.xlim(0, 20)
plt.legend()
plt.title(f'Binomial distribution for different values of n\na={a}')
```

```python
# now compare to stirling approximation (step 1)

def stirling_binomial(n, k, a):
    return np.sqrt(n/(2*np.pi*k*(n-k))) * n**n * a**k * (1 - a)**(n - k)/ (k**k * (n - k)**(n - k))
```

```python
# plot binomial distribution and stirling approx for different values of n= 10
n = 10
a=0.7
k_cont = np.linspace(0, n, 100)

plt.plot(k_values, binomial(n, k_values, a), label=f'binomial n={n}')
plt.plot(k_cont, stirling_binomial(n, k_cont, a), label=f'stirling n={n}')
plt.xlabel('k')
plt.ylabel('P(X=k)')
plt.legend()
plt.title(f'Binomial distribution and Stirling approximation for n={n}\na={a}')

```

```python
# almost gaussian

def almost_gaussian_binomial(n, k, a):
    return 1/np.sqrt(2*np.pi) * np.sqrt(n/(k*(n-k)))*np.exp(-((k-n*a)**2)/(2*n*a*(1-a)))
```

```python
# compare
n = 20
a = 0.3
k_cont = np.linspace(0, n, 100)

plt.plot(k_values, binomial(n, k_values, a), label=f'binomial n={n}')
plt.plot(k_cont, almost_gaussian_binomial(n, k_cont, a), label=f'almost gaussian n={n}')
plt.plot(k_cont, stirling_binomial(n, k_cont, a), label=f'stirling n={n}')
plt.xlabel('k')
plt.ylabel('P(X=k)')
plt.legend()
plt.title(f'Binomial distribution and almost Gaussian approximation for n={n}\na={a}')
```

```python
# now actually gaussian

def gaussian_binomial(n, k, a):
    return 1/np.sqrt(2*np.pi*n*a*(1-a)) * np.exp(-((k-n*a)**2)/(2*n*a*(1-a)))

```

```python
# compare
n = 10
a = 0.2
k_cont = np.linspace(0, n, 100)

plt.plot(k_values, binomial(n, k_values, a), label=f'binomial n={n}')
plt.plot(k_cont, gaussian_binomial(n, k_cont, a), label=f'gaussian n={n}')
plt.plot(k_cont, almost_gaussian_binomial(n, k_cont, a), label=f'almost gaussian n={n}')
plt.plot(k_cont, stirling_binomial(n, k_cont, a), label=f'stirling n={n}')
plt.xlabel('k')
plt.ylabel('P(X=k)')
plt.legend()
```
