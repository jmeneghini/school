c) Find the Fourier series for $f (x)$.

```python
x = Symbol('x')
n = Symbol('n', integer=True)
```

```python
f = Abs(sin(x))
```

```python
fourier = fourier_series(f, (x, -pi, pi))
```

```python
plot(fourier.truncate(18), (x, -pi, pi), ylim=(0, 1), xlabel='$x$', ylabel='$f(x)$')
```
