```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *
```

```python
x = np.linspace(-np.pi, np.pi, 100)
```

```python
def fourier(num_terms, x):
    a_0 = 2/np.pi
    for i in range(1, num_terms+1):
        if i % 2 == 0:
            a_n = 2/np.pi*(1/(1+i) + 1/(1-i))
        else:
            a_n = 0
        a_0 += a_n*np.cos(i*x)
    return a_0
```

```python
plt.plot(x, fourier(2, x), label='n=2')
plt.plot(x, fourier(4, x), label='n=8')
plt.plot(x, fourier(8, x), label='n=8')
plt.plot(x, fourier(12, x), label='n=12')
plt.plot(x, fourier(16, x), label='n=16')
plt.plot(x, abs(np.sin(x)), label='|sin(x)|')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.savefig("HW3.png", dpi=300)
```
